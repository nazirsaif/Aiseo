# Multi-Label SEO Issue Dataset Generator
This notebook synthetically generates 10,000+ realistic SEO data points relating 16 technical features to 13 distinct, identifiable issues.

```python
import pandas as pd
import numpy as np
import os

np.random.seed(42)
NUM_SAMPLES = 500000
```

```python
# CELL 2: Generate 16 Core Features
data = {
    'content_length': np.random.randint(50, 3000, NUM_SAMPLES),
    'keyword_density': np.random.uniform(0.1, 5.0, NUM_SAMPLES),
    'num_internal_links': np.random.randint(0, 50, NUM_SAMPLES),
    'num_external_links': np.random.randint(0, 20, NUM_SAMPLES),
    'has_meta_description': np.random.choice([0, 1], NUM_SAMPLES, p=[0.3, 0.7]),
    'has_alt_text': np.random.choice([0, 1], NUM_SAMPLES, p=[0.4, 0.6]),
    'avg_time_on_page_sec': np.random.randint(5, 300, NUM_SAMPLES),
    'bounce_rate': np.random.uniform(20.0, 95.0, NUM_SAMPLES),
    'scroll_depth_percent': np.random.uniform(10.0, 100.0, NUM_SAMPLES),
    'domain_authority': np.random.randint(1, 100, NUM_SAMPLES),
    'page_authority': np.random.randint(1, 100, NUM_SAMPLES),
    'backlink_count': np.random.randint(0, 500, NUM_SAMPLES),
    'serp_position_before': np.random.randint(1, 100, NUM_SAMPLES),
    # --- 3 NEW FEATURES POST-ARCHITECTURAL SHIFT ---
    'h1Count': np.random.randint(0, 4, NUM_SAMPLES), # Pages randomly have 0, 1, 2, or 3 H1s
    'has_readable_font_size': np.random.choice([0, 1], NUM_SAMPLES, p=[0.15, 0.85]), # Minor chance of bad font size
    'isNoindex': np.random.choice([0, 1], NUM_SAMPLES, p=[0.9, 0.1]) # Minor chance of accidental no-index block
}
```

```python
# CELL 3: Map Features to 13 Ground-Truth Technical Issues (Labels)
df = pd.DataFrame(data)

# A page has an issue if it hits specific strict conditions
df['issue_low_content'] = (df['content_length'] < 300).astype(int)
df['issue_missing_meta'] = (df['has_meta_description'] == 0).astype(int)
df['issue_missing_alt'] = (df['has_alt_text'] == 0).astype(int)
df['issue_no_internal_links'] = (df['num_internal_links'] == 0).astype(int)
df['issue_no_external_links'] = (df['num_external_links'] == 0).astype(int)
df['issue_poor_density'] = ((df['keyword_density'] < 0.5) | (df['keyword_density'] > 3.0)).astype(int)
df['issue_high_bounce_rate'] = (df['bounce_rate'] > 75.0).astype(int)
df['issue_low_time'] = (df['avg_time_on_page_sec'] < 30).astype(int)
df['issue_low_scroll'] = (df['scroll_depth_percent'] < 40.0).astype(int)
df['issue_low_authority'] = (df['domain_authority'] < 20).astype(int)

# --- 3 NEW TARGET ISSUES MAP TO NEW FEATURES ---
df['issue_no_h1'] = (df['h1Count'] != 1).astype(int)              # Issue if 0 OR multiple H1s!
df['issue_bad_font_size'] = (df['has_readable_font_size'] == 0).astype(int)
df['issue_is_noindex'] = (df['isNoindex'] == 1).astype(int)
```

```python
# CELL 4: Evaluate overall theoretical Compliance score out of 100 based on issues
total_issues = df[['issue_low_content', 'issue_missing_meta', 'issue_missing_alt', 
                   'issue_no_internal_links', 'issue_no_external_links', 'issue_poor_density', 
                   'issue_high_bounce_rate', 'issue_low_time', 'issue_low_scroll', 
                   'issue_low_authority', 'issue_no_h1', 'issue_bad_font_size', 'issue_is_noindex']].sum(axis=1)

# Score starts at 100, drops by approx 8 for every issue
df['pytorch_compliance_score'] = np.clip(100 - (total_issues * 8), 0, 100)

print(df.head())
```

```python
# CELL 5: Export to CSV
os.makedirs('Dataset', exist_ok=True)
df.to_csv('Dataset/seo_multi_issues_dataset.csv', index=False)
print("Massive Synthetic Dataset Generated: Dataset/seo_multi_issues_dataset.csv")
```
