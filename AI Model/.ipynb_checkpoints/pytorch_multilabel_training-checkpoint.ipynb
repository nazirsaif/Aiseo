# Deep Multi-Label Issue Predictor (PyTorch)
This trained deep network identifies SEO issues accurately across 16 structural metrics. The network utilizes deep hidden layers with Batch Normalization and Dropout for extreme regularization against massive scaling requirements.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib

print(f"PyTorch Version: {torch.__version__}")
```

```python
# CELL 2: ARCHITECTURE (DEEP NEURAL NETWORK)
class SEOMultiIssuePredictor(nn.Module):
    def __init__(self, input_size, num_issues):
        super(SEOMultiIssuePredictor, self).__init__()
        
        # Deep Hidden Layer Architecture (Expanded for better prediction mapping)
        self.layer1 = nn.Linear(input_size, 64)
        self.bn1 = nn.BatchNorm1d(64)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(p=0.2)
        
        self.layer2 = nn.Linear(64, 32)
        self.bn2 = nn.BatchNorm1d(32)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(p=0.2)
        
        self.layer3 = nn.Linear(32, 16)
        self.relu3 = nn.ReLU()
        
        # Multi-Label Head (Issues)
        self.out_issues = nn.Linear(16, num_issues)
        # Scoring Head (1 Final Score)
        self.out_score = nn.Linear(16, 1)
        
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # Layer 1
        x = self.layer1(x)
        x = self.bn1(x)
        x = self.relu1(x)
        x = self.dropout1(x)
        
        # Layer 2
        x = self.layer2(x)
        x = self.bn2(x)
        x = self.relu2(x)
        x = self.dropout2(x)
        
        # Layer 3
        x = self.layer3(x)
        x = self.relu3(x)
        
        # Parallel Extraction
        issues = self.sigmoid(self.out_issues(x))
        score = self.sigmoid(self.out_score(x))
        
        return issues, score

print("Deep Learning Architecture Established.")
```

```python
# CELL 3: DATA PREPARATION & SCALING (Updated with DataLoaders)
df = pd.read_csv('Dataset/seo_multi_issues_dataset.csv')

# 16 Base Features
feature_cols = ['content_length', 'keyword_density', 'num_internal_links', 'num_external_links', 
                'has_meta_description', 'has_alt_text', 'avg_time_on_page_sec', 'bounce_rate',
                'scroll_depth_percent', 'domain_authority', 'page_authority', 
                'backlink_count', 'serp_position_before', 
                'h1Count', 'has_readable_font_size', 'isNoindex']

# 13 Technical Issue Labels
issue_cols = ['issue_low_content', 'issue_missing_meta', 'issue_missing_alt', 
              'issue_no_internal_links', 'issue_no_external_links', 'issue_poor_density', 
              'issue_high_bounce_rate', 'issue_low_time', 'issue_low_scroll', 
              'issue_low_authority', 'issue_no_h1', 'issue_bad_font_size', 'issue_is_noindex']

X = df[feature_cols].values
y_issues = df[issue_cols].values
y_score = (df['pytorch_compliance_score'] / 100.0).values.reshape(-1, 1)

X_train, X_test, y_issues_train, y_issues_test, y_score_train, y_score_test = train_test_split(
    X, y_issues, y_score, test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_t = torch.FloatTensor(X_train_scaled)
y_issues_train_t = torch.FloatTensor(y_issues_train)
y_score_train_t = torch.FloatTensor(y_score_train)

X_test_t = torch.FloatTensor(X_test_scaled)
y_issues_test_t = torch.FloatTensor(y_issues_test)
y_score_test_t = torch.FloatTensor(y_score_test)

from torch.utils.data import TensorDataset, DataLoader

# Create PyTorch Datasets
train_dataset = TensorDataset(X_train_t, y_issues_train_t, y_score_train_t)
test_dataset = TensorDataset(X_test_t, y_issues_test_t, y_score_test_t)

# Create DataLoaders (Using 1024 batch size for the larger 400k dataset to optimize training speed)
batch_size = 1024
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print(f"DataLoaders created. {len(train_loader)} training batches per epoch.")
```

```python
# CELL 4: DEEP LEARNING TRAINING PROCESS
import copy
from sklearn.metrics import classification_report
import io

model = SEOMultiIssuePredictor(input_size=len(feature_cols), num_issues=len(issue_cols))

# Loss & Optimizer
criterion_issues = nn.BCELoss() 
criterion_score = nn.MSELoss()  
optimizer = optim.Adam(model.parameters(), lr=0.005, weight_decay=1e-5)

# IMPROVEMENT: Learning Rate Scheduler
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=20)

epochs = 2000 
history = {'loss': [], 'val_loss': [], 'train_accuracy': [], 'val_accuracy': []}

def extract_accuracy(preds, labels):
    binary_preds = (preds > 0.5).float()
    correct = (binary_preds == labels).float().sum()
    return (correct / (labels.size(0) * labels.size(1))) * 100.0

# IMPROVEMENT: Early Stopping Setup
best_val_loss = float('inf')
patience_counter = 0
early_stop_patience = 50
best_model_weights = copy.deepcopy(model.state_dict())

for epoch in range(epochs):
    # --- TRAINING PHASE ---
    model.train()
    running_train_loss = 0.0
    running_train_acc = 0.0
    
    for batch_x, batch_y_issues, batch_y_score in train_loader:
        optimizer.zero_grad()
        
        out_issues, out_score = model(batch_x)
        loss_issues = criterion_issues(out_issues, batch_y_issues)
        loss_score = criterion_score(out_score, batch_y_score)
        
        total_loss = loss_issues + (loss_score * 0.5)
        total_loss.backward()
        optimizer.step()
        
        running_train_loss += total_loss.item() * batch_x.size(0)
        running_train_acc += extract_accuracy(out_issues, batch_y_issues).item() * batch_x.size(0)
        
    epoch_train_loss = running_train_loss / len(train_loader.dataset)
    epoch_train_acc = running_train_acc / len(train_loader.dataset)
    history['loss'].append(epoch_train_loss)
    history['train_accuracy'].append(epoch_train_acc)
    
    # --- VALIDATION PHASE ---
    model.eval()
    running_val_loss = 0.0
    running_val_acc = 0.0
    
    with torch.no_grad():
        for batch_x, batch_y_issues, batch_y_score in test_loader:
            val_issues, val_score = model(batch_x)
            v_loss_issues = criterion_issues(val_issues, batch_y_issues)
            v_loss_score = criterion_score(val_score, batch_y_score)
            v_total_loss = v_loss_issues + (v_loss_score * 0.5)
            
            running_val_loss += v_total_loss.item() * batch_x.size(0)
            running_val_acc += extract_accuracy(val_issues, batch_y_issues).item() * batch_x.size(0)
            
    epoch_val_loss = running_val_loss / len(test_loader.dataset)
    epoch_val_acc = running_val_acc / len(test_loader.dataset)
    history['val_loss'].append(epoch_val_loss)
    history['val_accuracy'].append((epoch, epoch_val_acc))
    
    # Trigger Scheduler
    scheduler.step(epoch_val_loss)
    
    # --- EARLY STOPPING CHECK ---
    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        best_model_weights = copy.deepcopy(model.state_dict())
        patience_counter = 0
    else:
        patience_counter += 1
        
    if patience_counter >= early_stop_patience:
        print(f"\nEarly stopping triggered at epoch {epoch+1}! Restoring best weights.")
        model.load_state_dict(best_model_weights)
        break
        
    # Print progress every 10 epochs
    if (epoch + 1) % 10 == 0 or epoch == 0:
        print(f"Epoch [{(epoch + 1):4d}/{epochs}] | Train Loss: {epoch_train_loss:.4f} | Val Loss: {epoch_val_loss:.4f} | Train Acc: {epoch_train_acc:.2f}% | Val Acc: {epoch_val_acc:.2f}%")

print("\nModel Training Optimized & Finished!")

# --- FINAL EVALUATION ON TEST DATASET ---
model.eval()
all_preds = []
all_labels = []
with torch.no_grad():
    for batch_x, batch_y_issues, _ in test_loader:
        val_issues, _ = model(batch_x)
        binary_preds = (val_issues > 0.5).int()
        all_preds.append(binary_preds.numpy())
        all_labels.append(batch_y_issues.int().numpy())

all_preds = np.vstack(all_preds)
all_labels = np.vstack(all_labels)

report_str = classification_report(all_labels, all_preds, target_names=issue_cols, zero_division=0)
print("\n--- TEST DATASET EVALUATION REPORT ---")
print(report_str)

# Write report to file to persist it
with open('metrics_report.txt', 'w') as f:
    f.write("--- TEST DATASET EVALUATION REPORT ---\n")
    f.write(report_str)
```

```python
# CELL 5: GENERATE ANALYTICS GRAPHS
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1. Loss Graph
axes[0].plot(history['loss'], color='red', linewidth=1.5, label='Train Loss')
axes[0].plot(history['val_loss'], color='orange', linewidth=1.5, label='Val Loss')
axes[0].set_title('Neural Network Training & Validation Loss')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Composite Loss (BCE + MSE)')
axes[0].legend()
axes[0].grid(True, linestyle='--', alpha=0.6)

# 2. Accuracy Graph
axes[1].plot(history['train_accuracy'], color='blue', label='Train Accuracy', alpha=0.7)
val_epochs, val_accs = zip(*history['val_accuracy'])
axes[1].plot(val_epochs, val_accs, color='green', marker='o', label='Val Accuracy')
axes[1].set_title('Prediction Accuracy Scale')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Accuracy %')
axes[1].legend()
axes[1].grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.savefig('Dataset/training_performance.png')
plt.show()
```

```python
# CELL 6: EXPORT COMPILED WEIGHTS
model.eval() # Secure into inference state
torch.save(model.state_dict(), 'seo_issues_classifier.pth')
joblib.dump(scaler, 'seo_issues_scaler.pkl')
joblib.dump(feature_cols, 'seo_issues_features.pkl')
joblib.dump(issue_cols, 'seo_issues_cols.pkl')

print("\nDeep Neural Engine Weights successfully exported for API integration.")
```
